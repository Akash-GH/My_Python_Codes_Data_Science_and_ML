#### Heatmap

##### For matrix plot the data should be in matrix form. So here inorder to convert it into matrix we have get the correlated data

```python
tips.corr(numeric_only = True)
```

```python
cd = tips.corr(numeric_only = True)
```

```python
sns.heatmap(cd)
```

```python
# If True, write the data value in each cell. If an array-like with the same shape as ``data``, then use this to annotate 
# the heatmap instead of the data. Note that DataFrames will match on position, not index.
sns.heatmap(cd, annot = True)
```

```python
sns.heatmap(cd, annot = True, cmap = 'coolwarm')
```

```python
flights
```

```python
flights.pivot_table(values = 'passengers', index = 'month',columns = 'year')
```

```python
matrixform = flights.pivot_table(values = 'passengers', index = 'month',columns = 'year')
```

```python
sns.heatmap(matrixform)
```

```python
sns.heatmap(matrixform,cmap = 'magma')
```

```python
sns.heatmap(matrixform,cmap = 'magma',linewidth = 1)
```

```python
sns.heatmap(matrixform,cmap = 'coolwarm',linewidth = 1)
```

#### Clustermap
##### The clustermap uses hierarchal clustering to produce a clustered version of the heatmap.

```python
sns.clustermap(matrixform)
```

```python
sns.clustermap(matrixform,standard_scale = 1)
```
