```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
iris.head()
```

#### pairplot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

#### PairGrid

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
# mapped the whole in scatterplot form
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# mapping diagonal,upper and lower in different plots
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot,kde = True)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

#### FacetGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(sns.histplot, "total_bill", kde = True)
```

```python
sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
```

```python
# This is an example of when we need two arguments for doing a particular plot
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
g.map(plt.scatter, "total_bill", "tip").add_legend()
```
