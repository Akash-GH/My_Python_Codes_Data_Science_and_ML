```python
X = df.drop('benign_0__mal_1',axis = 1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30, activation= 'relu'))

model.add(Dense(units = 15, activation= 'relu'))

# sigmoid due to it is a binary classfication
model.add(Dense(units = 1, activation= 'sigmoid'))

model.compile(optimizer='adam', loss= 'binary_crossentropy')
```

```python
model.fit(x = X_train, y = y_train, epochs=600, validation_data=(X_test,y_test))
```

```python
model.history.history
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.head()
```

```python
# This is the perfect example of overfitting
losses.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30, activation= 'relu'))

model.add(Dense(units = 15, activation= 'relu'))

# sigmoid due to it is a binary classfication
model.add(Dense(units = 1, activation= 'sigmoid'))

model.compile(optimizer='adam', loss= 'binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 25)
```

```python
model.fit(x = X_train, y = y_train, epochs=600, validation_data=(X_test,y_test), callbacks = [early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30, activation= 'relu'))
model.add(Dropout(rate=0.5))# This will make probabilitiy of turning off 50% of neurons

model.add(Dense(units = 15, activation= 'relu'))
model.add(Dropout(rate=0.5))# This will make probabilitiy of turning off 50% of neurons

# sigmoid due to it is a binary classfication
model.add(Dense(units = 1, activation= 'sigmoid'))

model.compile(optimizer='adam', loss= 'binary_crossentropy')
```

```python
model.fit(x = X_train, y = y_train, epochs=600, validation_data=(X_test,y_test), callbacks = [early_stop])
```

```python
dmodel_loss = pd.DataFrame(model.history.history)
dmodel_loss.plot()
```
